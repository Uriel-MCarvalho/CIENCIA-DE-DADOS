# Experimento 03


Neste experimento, investigaremos colisões em um cenário envolvendo um carro em movimento em um trilho de ar inclinado equipado com uma mola. O objetivo é caracterizar a colisão, observando os eventos que ocorrem antes, durante e após o impacto.

> *_O conceito de colisão em Física vai além de simples choques entre objetos sólidos; abrange qualquer interação entre corpos em movimento relativo. Esse tópico é fundamental em várias áreas da Física, desde o choque de bolas de bilhar até colisões subatômicas, visando entender as forças envolvidas durante a colisão._*


As colisões são classificadas como elásticas ou inelásticas, dependendo da perda de Energia Mecânica durante o processo. A elasticidade é quantificada pelo coeficiente de restituição (ε), onde $ε = 1$ representa uma colisão perfeitamente elástica, e $ε = 0$, uma colisão perfeitamente inelástica.
$\displaystyle ε = \frac{v'}{v}$
Neste experimento, se a colisão entre o carro e a mola for elástica $(0 < ε < 1)$, o carro retornará a uma posição menor que a de soltura ao longo do trilho. Isso pode ser descrito por uma relação exponencial do tipo $Y = C e^{-nX}$, que é comumente observada em diversos fenômenos físicos.

Além de determinar o coeficiente de restituição em colisões inelásticas, utilizaremos gráficos em escala logarítmica como ferramenta de análise da função exponencial.

Agora, vamos prosseguir com a descrição detalhada do experimento e dos resultados obtidos.


>**Objetivos**

Neste experimento, o principal objetivo é investigar o comportamento das colisões entre um carrinho em movimento e uma mola localizada na base de um trilho de ar inclinado. Especificamente, pretendemos:

3. Analisar a variação da energia mecânica durante as diferentes colisões.

3. Preparação do elástico na base do trilho (6 tiras).

4. Ajuste da unidade de fluxo de ar.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

def plot_collision_distance(df_collisions, series_labels=None, subset_size=None):
    # Configurar estilo do Seaborn
    sns.set_theme(style="darkgrid")

    # Se series_labels não for fornecido, use todas as colunas do DataFrame
    if series_labels is None:
        series_labels = df_collisions.columns[1:]

    # Plotar as linhas para as diferentes séries (colunas) do DataFrame
    plt.figure(figsize=(10, 6))
    for label in series_labels:
        sns.lineplot(data=df_collisions, x='Nº Colisões', y=label, label=label, marker='o')

        # Ajustar um modelo de regressão linear
        model = LinearRegression()
        X = np.log(df_collisions['Nº Colisões'].values.reshape(-1, 1))
        y = df_collisions[label].values
        model.fit(X, y)

        # Obter coeficientes da regressão
        slope = model.coef_[0]
        intercept = model.intercept_

        print(f"Para {label}: Coeficiente angular (slope): {slope}, Intercepto: {intercept}")

    # Selecionar um subconjunto dos dados se especificado
    if subset_size:
        df_subset = df_collisions.head(subset_size)
        sns.regplot(x='Nº Colisões', y=series_label, data=df_subset, scatter=False, color='r')
    else:
        sns.regplot(x='Nº Colisões', y=series_label, data=df_collisions, scatter=False, color='r')

    # Definir rótulo dos eixos
    plt.xlabel('Nº Colisões')
    plt.ylabel('Distância (cm)')

    # Definir escala logarítmica no eixo X
    plt.xscale('log')

    # Definir título
    plt.title('Distância X N° Colisões (Escala Logarítmica)')

    # Exibir o gráfico
    plt.legend(title='Volume', loc='upper right')
    plt.show()

# Exemplo de uso
plot_collision_distance(df_collisions, subset_size=3)
```

#### Grafíco 03

Neste gráfico é apresentado o desvio padrão entre cada medida por colisão. Nos gráficos anteriores, observa-se uma pequena diferença que aparentemente não tem relação com a massa, então as medidas foram verificadas.

```python
# Configurar o estilo do Seaborn
sns.set_theme(style="ticks")

# Supondo que df_collisions seja o seu DataFrame
# Variável desvio_padrao_por_linha
desvio_padrao_por_linha = df_collisions[['0V', '1/4V', '2/4V', '3/4V', '4/4V']].std(axis=1)

# Inicializar a figura com um eixo x logarítmico
f, ax = plt.subplots(figsize=(7, 6))

# Plotar o desvio_padrao_por_linha com caixas horizontais
sns.boxplot(x='Nº Colisões', y=desvio_padrao_por_linha, data=df_collisions,
            whis=[0, 100], width=.6, palette="vlag")

# Adicionar pontos para mostrar cada observação
sns.stripplot(x='Nº Colisões', y=desvio_padrao_por_linha, data=df_collisions,
              size=4, color=".3", linewidth=0)

# Adicionar uma linha de regressão linear
sns.regplot(x='Nº Colisões', y=desvio_padrao_por_linha, data=df_collisions,
            scatter=False, color='r')

# Calcular a inclinação da linha de regressão
coefficients = np.polyfit(df_collisions['Nº Colisões'], desvio_padrao_por_linha, 1)
slope = coefficients[0]

# Exibir a inclinação
print(f"Inclinação da Regressão Linear: {slope}")

# Ajustar a apresentação visual
ax.xaxis.grid(True)
ax.set(ylabel="")
sns.despine(trim=True, left=True)

# Mostrar o gráfico
plt.show()
print(slope)
```


Considerando o desvio padrão, pode-se afirmar que o erro observado anteriormente ocorre devido às medidas realizadas, uma vez que uma medida não está muito distante da outra.
